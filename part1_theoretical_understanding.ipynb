## Part 1: Theoretical Understanding (40%)

### Q1: Explain the primary differences between TensorFlow and PyTorch. When would you choose one over the other?
**Answer:**
- **TensorFlow** uses static computation graphs (with eager mode optional) and is production-oriented, ideal for deployment and mobile apps (with TFLite).
- **PyTorch** uses dynamic computation graphs, which makes it more flexible and easier to debug.
- Choose **TensorFlow** for scalability and production, and **PyTorch** for research, prototyping, and when flexibility is important.

### Q2: Describe two use cases for Jupyter Notebooks in AI development.
**Answer:**
1. **Exploratory data analysis (EDA):** Visualizing data distributions, relationships, and cleaning steps.
2. **Model training and documentation:** Writing and testing ML/DL code with markdown explanations for reproducibility.

### Q3: How does spaCy enhance NLP tasks compared to basic Python string operations?
**Answer:**
- spaCy uses pre-trained models and efficient tokenization for linguistic features like POS tagging, NER, and dependency parsing.
- It is faster and more accurate than using `split()` or regular expressions, and supports multi-language NLP pipelines.

## 2. Comparative Analysis: Scikit-learn vs TensorFlow
| Feature | Scikit-learn | TensorFlow |
|--------|---------------|------------|
| Target Use | Classical ML (SVM, trees) | Deep Learning (NNs, CNNs, RNNs) |
| Beginner Friendly | Very easy to use | Steeper learning curve |
| Community | Strong, academic & applied | Massive, industry & research |
| Deployment | Limited support | Excellent (TFLite, TF Serving) |
